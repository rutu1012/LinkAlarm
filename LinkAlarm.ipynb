```python
import datetime

def alarm(hour,minute,am,day_sch,meetLinks):
    n = len(hour) #no. of lectures for the day

    for i in range(n):
        hour[i] = int(hour[i])
        minute[i] = int(minute[i])
        if am[i] == "PM" and hour[i] != 12:
            hour[i] = 12 + int(hour[i])

        

        while True:
            currentTime = datetime.datetime.utcnow()+datetime.timedelta(hours=5,minutes=30) #adjusted to India's timezone

            if currentTime.hour > hour[i] :
                print(list(day_sch.keys())[i]+" lec time has been passed ")
                break
            elif currentTime.hour == hour[i] and currentTime.minute > minute[i]+30:
                print(list(day_sch.keys())[i]+" lec time has been passed ")
                break
            
            elif currentTime.hour == hour[i] and currentTime.minute == minute[i]:
                clear_output()
                print(currentTime)
                print("Class time now!!!")
                    
                print("\nLecture : "+list(day_sch.keys())[i])
                print("Meeting Link: "+meetLinks[i])
                print()
                playAudio()
                time.sleep(60) 
                break
                
            elif currentTime.hour == hour[i] and currentTime.minute < minute[i]+30:
                print(list(day_sch.keys())[i]+" lec has already started Join Fast! ")
                print("Meeting Link: "+meetLinks[i])
                playAudio()
                break
```

```python
def time_split(day_sch): 
    n = len(day_sch)
    l, hour, minute ,am= [],[],[],[]
    j = 0
    while j < n:
        for i in day_sch:
            l.append(day_sch[i].split('-')[0].split(' '))
            j+=1

    for i in range(n):
        hour.append(l[i][0].split(':')[0])
        minute.append(l[i][0].split(':')[1])
        am.append(l[i][1])

    return hour,minute,am
```

```python
def sorting(schedule):
    hr,min,am_pm = time_split(schedule)
    t_sch=[]
    for i in range(len(schedule.values())):
        if int(hr[i])<10:
            hr[i]='0'+hr[i]
        t_sch.append(hr[i]+":"+min[i]+' '+am_pm[i])
    
    for k in schedule.keys():
        l=schedule[k].split(" ")
        h,m=l[0].split(":")
        if int(h)<10:
            h1=h
            hr='0'+h
            for i in range(len(schedule[k])):
                if schedule[k].find(h1)!=-1:
                    l=list(schedule[k])
                    l[i]=hr
                    schedule[k]=''.join(l)
                    break

    
    format = '%I:%M %p'
    time_hours = [time.strptime(t,format) for t in t_sch]
    sorted_time = [time.strftime(format,h) for h in sorted(time_hours)]
    new_sch={}
    sort_link=[]
    for i in range(len(sorted_time)):
        for k in schedule.keys():
            if schedule[k].find(sorted_time[i])!=-1:
                new_sch[k]=schedule[k]
                sort_link.append(meetLink[i])

    l=[]

    for i in range(len(sorted_time)):
        l.append(sorted_time[i].split(' '))
        

    hour,minute,am=[],[],[]
    
    for i in range(len(l)):
        
        am.append(l[i][1])
        hour.append(l[i][0].split(':')[0])
        minute.append(l[i][0].split(':')[1])
        
    return hour,minute,am,new_sch,sort_link
```

```python
hr,min,am_pm,sort_sch,sort_link=sorting(schedule) # first sort the schedule input timings
print(tt(sort_sch))
print()
alarm(hr,min,am_pm,sort_sch,sort_link)
```
